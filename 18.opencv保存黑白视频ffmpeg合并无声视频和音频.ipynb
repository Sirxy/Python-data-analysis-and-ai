```python
import numpy as np

import cv2
```

```python
cap = cv2.VideoCapture("./mp4/stars.mp4")

while True:
    
    flag, frame = cap.read()
    
    gray = cv2.cvtColor(frame, code=cv2.COLOR_BGR2GRAY)
    print(gray.shape)
    
    cv2.imshow("video", gray)
    
    if ord("q") == cv2.waitKey(40):
        break
    
cap.release()
cv2.destroyAllWindows()
```
