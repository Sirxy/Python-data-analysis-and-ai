```python
# 加载图片准备
import matplotlib.pyplot as plt
# jupyter魔法指令，预加载，显示图片
%matplotlib inline
```

```python
img = plt.imread('./maomi.jpg')
img  # 所有的图片其实都是numpy.ndarray.数组对象，彩色图片是三维的数组对象，长度宽度颜色
```

```python
plt.imshow(img)
```

```python
# 查看数组形状
img.shape
```

```python
# 将图片的一部分切出来了
img2 = img[300:700, 0:300]
plt.imshow(img2)
```

```python
img2 = img[::-1]
plt.imshow(img2)
```

```python
# 彩色图片，三维，最后一维代表颜色，颜色默认红绿蓝，蓝绿红
img4 = img[:, :, ::-1]
plt.imshow(img4)
```

```python
rose = plt.imread('./red_rose.jpg')
plt.imshow(rose)
```

```python
rose2 = rose[:, :, ::-1]
plt.imshow(rose2)
```
