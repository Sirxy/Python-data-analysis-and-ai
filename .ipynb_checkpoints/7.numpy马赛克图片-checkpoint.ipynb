```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 切片，从图片中选取一部分
# 从数据中，每隔几个取出数据
# 对图片而言，如果间隔足够大，显示马赛克效果
```

```python
rose = plt.imread("./rose.jpg")
plt.imshow(rose)
```

```python
rose.shape
```

```python
rose2 = rose[::5, ::5]
plt.imshow(rose2)
```

```python
rose2.shape
```

```python
rose3 = rose[::10, ::10]
plt.imshow(rose3)
```

```python
jin = plt.imread('./jin.jpg')
plt.imshow(jin)
```

```python
head = jin[50:160, 200:280]
plt.imshow(head)
```

```python
head.shape
```

```python
jin2 = jin.copy()
jin2[50:160, 200:280] = rose[:110, :80]
plt.imshow(jin2)
```

```python
jin2[50:160, 200:280] = head
plt.imshow(jin2)
```

```python
head2 = head[::5,::5]
plt.imshow(head2)
```

```python
head2.shape
```

```python
# ValueError: could not broadcast input array from shape (22,16,3) into shape (110,80,3) 
# 即形状不匹配，head2 (22,16,3) ，jin2 (110,80,3)
jin2[50:160, 200:280] = head2
plt.imshow(jin2)
```

```python
for i in range(22):
    for j in range(16):
        jin2[50 + i*5: 50 + 5 + i*5, 200 + j*5: 200 + 5 + j * 5] = head2[i,j]
plt.imshow(jin2)
```
