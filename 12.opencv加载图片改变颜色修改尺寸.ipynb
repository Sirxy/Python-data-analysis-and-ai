```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

# cv2 模块下，都是使用Python方法调用c++方法，
import cv2
```

```python
rose = cv2.imread('./rose.jpg')  # cv2在读取与图片时，颜色通道蓝绿红
rose.shape
```

```python
plt.imshow(rose)
```

```python
plt.imshow(rose[:,:,::-1])
```

```python
# cv2也可以显示图片

gray = cv2.cvtColor(rose, code=cv2.COLOR_BGR2GRAY)

img = cv2.resize(gray, dsize=(200, 100))

cv2.imshow('rose', img)

# 等待键盘的输入
while True:
    if ord('q') == cv2.waitKey(100):
        break

# 释放内存
cv2.destroyAllWindows()
```
